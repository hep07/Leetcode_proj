Valid Tic-Tac-Toe State

```python
class Solution:
    def validTicTacToe(self, board):
        """
        :type board: List[str]
        :rtype: bool
        """
        row_val = [ 0 for i in range(len(board))]
        col_val = [ 0 for i in range(len(board))]
        diag_val = [0, 0]
        X_count = 0
        O_count = 0
        val_dict = {"X":1, "O":-1, " ":0}
        for i in range(len(board)):
            for j in range(len(board[i])):
                if i == j:
                    diag_val[0] += val_dict[board[i][j]]
                if i + j == len(board) - 1:
                    diag_val[1] += val_dict[board[i][j]]
                
                if board[i][j] == "X":
                    row_val[i] += 1
                    col_val[j] += 1
                    X_count += 1
                elif board[i][j] == "O":
                    row_val[i] += -1
                    col_val[j] += -1
                    O_count += 1
                    
        if X_count - O_count != 0 and X_count - O_count != 1:
            return False
        
        # If X wins, last move must be from X player. Therefore X_counts - O_counts == 1
        # On the other hand, if O wins, X_counts == O_counts
        X_win = any([x for x in row_val if x == 3]) or any([x for x in col_val if x == 3]) or any([x for x in diag_val if x == 3])
        Y_win = any([x for x in row_val if x == -3]) or any([x for x in col_val if x == -3]) or any([x for x in diag_val if x == -3])
        
        if X_win and X_count == O_count:
            return False
        
        if Y_win and X_count == O_count + 1:
            return False
        
        # The above two conditions guarunteed that if X_win and Y_win, we return False
        
        return True
```

LC 348 Design Tic Tac Toe

```python
class TicTacToe(object):

    def __init__(self, n):
        """
        Initialize your data structure here.
        :type n: int
        """
        #To achieve O(n), we need arrays to store the current value for each row, column and the two diagonal
        self.row_val = [ 0 for _ in range(n)]
        self.col_val = self.row_val.copy()
        self.diag_val = [0, 0]
        self.val_arr = [1, -1] # player 1 adds 1 to the row, col and diag, player 2 positions will add -1 to row, col and diag 
        self.n = n
        self.winning_check = [n, -n]
    def move(self, row, col, player):
        """
        Player {player} makes a move at ({row}, {col}).
        @param row The row of the board.
        @param col The column of the board.
        @param player The player, can be either 1 or 2.
        @return The current winning condition, can be either:
                0: No one wins.
                1: Player 1 wins.
                2: Player 2 wins.
        :type row: int
        :type col: int
        :type player: int
        :rtype: int
        """
        # majority of the work is in checking whether current player wins after making the move
        self.row_val[row]  += self.val_arr[player-1]
        self.col_val[col]  += self.val_arr[player-1]
        
        if row == col:
            self.diag_val[0] += self.val_arr[player-1]
        if row + col == self.n-1:
            self.diag_val[1] += self.val_arr[player-1]
            
        # check winning condition of current player. Since every move is valid, we only need to check whether there are winning conditions on current row, col and the two diag.
        # This is O(1) solution per move operation 
        cur_win = self.row_val[row] == self.winning_check[player-1] or self.col_val[col] == self.winning_check[player-1]  or any([x == self.winning_check[player-1] for x in self.diag_val])
        
        if cur_win:
            return player
        else:
            return 0
        

# Your TicTacToe object will be instantiated and called as such:
# obj = TicTacToe(n)
# param_1 = obj.move(row,col,player)
```
