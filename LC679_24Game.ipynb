```python
class Solution:
    def judgePoint24(self, nums):
        """
        :type nums: List[int]
        :rtype: bool
        """
        
        # It is tempting to try to use backtrack to traverse all 4!*4^3 possibilities
        # However that will not cover cases like a*b + c*d
        
        nums = nums.copy()
        tol = 1e-4
        
        def helper(arr):
            # when length of arr == 1, no more operations possible, we finish one possible calcualtion 
            if len(arr) ==1:
                # the following line is crucial, cases like 3 3 8 8 wont pass if we just check equality with 24
                return abs(arr[0] - 24) < tol
            
            # if there are more than 1 element, try all possible operations for any given pair, this will include case of a*b + c*d 
            # randomly select two numbers, add possible operations results to the arr
            for i in range(len(arr)-1):
                for j in range(i+1, len(arr)):
                                        
                    
                    next_possible = []
                    possible_next = [arr[j]-arr[i], 
                                     -arr[j]+arr[i],
                                     arr[j]+arr[i],
                                     arr[j]*arr[i]]
                    
                    if arr[j] != 0:
                        possible_next.append(arr[i]/arr[j])
                    
                    if arr[i] != 0:
                        possible_next.append(arr[j]/arr[i])
                    # backtrack idea, remove i,j from the arr
                    # j is the bigger index, so pop j first and then i 
                    b = arr.pop(j)
                    a = arr.pop(i)
                    # next push possible next_val
                    for next_val in possible_next:
                        arr.append(next_val)
                        if helper(arr):
                            return True                        
                        arr.pop()
                        
                    # adds back ith and jth element to its original position 
                    # adds position i first
                    arr.insert(i, a)
                    arr.insert(j, b)
            return False
        return helper(nums) 
                    
       
            














```
