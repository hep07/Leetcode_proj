```python
class Solution:
    def numIslands(self, grid):
        """
        :type grid: List[List[str]]
        :rtype: int
        """
        def edges(i,j, n,m):
            return list(filter(lambda node: node[0] >= 0 and node[0] < n and node[1] >=0 and node[1] < m, [ (i-1,j), (i+1,j), (i,j+1), (i,j-1)  ] ) )
    
        def dfs(i,j):
            stack = [(i,j)]
            while stack:
                this_i, this_j = stack.pop()
                for next_node in edges(this_i,this_j, n,m):
                    if grid[next_node[0]][next_node[1]] == "1" and next_node not in visited:
                        stack.append(next_node)
                        visited.add(next_node)
        
        # in this case fastest solution for this case should be graph search since we only query it at the end 
        # this will have o(nm) complexity
        
        n = len(grid)        
        if n ==0:
            return 0        
        m = len(grid[0])
        if m == 0:
            return 0 
        
        # convert to 1d index i * m + j 
        visited = set()
        count = 0
        for i in range(len(grid)):
            for j in range(len(grid[0])):
                if (i,j) not in visited and grid[i][j] == "1":
                    dfs(i,j) # this will mark all the visited nodes in visited set
                    count += 1
        
        return count
        
    
```

```python
class UnionFind:
    def __init__(self, n):
        self.size = [0 for _ in range(n)]
        self.parent = [x for x in range(n)]
        self.count = n
        
    def root_find(self, x):
        while self.parent[x]!=x:
            x= self.parent[x]            
        return x
    
    def union(self, a,b):
        root_a = self.root_find(a)
        root_b = self.root_find(b)
        
        if root_a != root_b:
            self.count -= 1
            if self.size[root_a] <= self.size[root_b]:
                self.parent[root_a] = root_b
                self.size[root_b] += self.size[root_a]
            else:
                self.parent[root_b] = root_a
                self.size[root_a] += self.size[root_b]
    
    def find(self, a,b):
        root_a = self.root_find(a)
        root_b = self.root_find(b)
        
        return root_a == root_b
        
        

class Solution:
    def numIslands2(self, m, n, positions):
        """
        :type m: int
        :type n: int
        :type positions: List[List[int]]
        :rtype: List[int]
        """
        # assume when calling edges, i,j are "1"
        
        marked_site = set()
        
        def edges(i,j, m,n):
            return list(filter(lambda node: node[0] >= 0 and node[0] < m and node[1] >=0 and node[1] < n and node[0]*n +node[1] in marked_site, [ (i-1,j), (i+1,j), (i,j+1), (i,j-1)  ] ) )
        
        if len(positions) == 0:
            return 0
        
        total_nodes = m*n
        myUF = UnionFind(total_nodes)
        
        res = []
        for pos in positions:
            marked_site.add( pos[0] * n + pos[1] )
            #print(marked_site)
            for adj in edges(pos[0], pos[1], m, n):
                #print(adj)
                myUF.union( pos[0] * n + pos[1], adj[0] * n + adj[1])
            res.append( len(marked_site) - ( total_nodes - myUF.count ) )# total lands thus far - ( num of unions performed thus far  ) = number of islands thus far
            
        return res
        
```
