```python
from collections import deque
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.stackin = []
        self.stackout = []
    def moveInToOut(self):
        # check that if stackout is empty, then pop all stackin items and push them onto stackout, essentially achieving reverse order here
        # only move from stackin to stackout WHEN stackout is EMPTY
        if len(self.stackout) ==0 :
            while self.stackin:
                self.stackout.append(self.stackin.pop())

    def push(self, x):
        """
        Push element x to the back of queue.
        :type x: int
        :rtype: void
        """
        self.stackin.append(x)

    def pop(self):
        """
        Removes the element from in front of queue and returns that element.
        :rtype: int
        """
        self.moveInToOut()
        return self.stackout.pop()

    def peek(self):
        """
        Get the front element.
        :rtype: int
        """
        self.moveInToOut()
        return self.stackout[-1]

    def empty(self):
        """
        Returns whether the queue is empty.
        :rtype: bool
        """
        return len(self.stackin)==0 and len(self.stackout) == 0


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

```python
from collections import deque
class MyStack:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue = deque()

    def push(self, x):
        """
        Push element x onto stack.
        :type x: int
        :rtype: void
        """
        self.queue.append(x)
        if len(self.queue) > 1:
            # move x towards the front like a bubble up operation
            for j in range(len(self.queue)-1, 0, -1):
                self.queue[j-1], self.queue[j] = self.queue[j], self.queue[j-1]
        # push operation is o(n) worst case
        

    def pop(self):
        """
        Removes the element on top of the stack and returns that element.
        :rtype: int
        """
        return self.queue.popleft()

    def top(self):
        """
        Get the top element.
        :rtype: int
        """
        return self.queue[0]

    def empty(self):
        """
        Returns whether the stack is empty.
        :rtype: bool
        """
        return len(self.queue) == 0


# Your MyStack object will be instantiated and called as such:
# obj = MyStack()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.top()
# param_4 = obj.empty()





```
