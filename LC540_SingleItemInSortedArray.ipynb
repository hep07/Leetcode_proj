```python
class Solution:
    def singleNonDuplicate_v1(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        lo = 0
        hi = len(nums) - 1
        if len(nums) == 1:
            return nums[0]
        
        if nums[-1] > nums[-2]:
            return nums[-1]
        
        if nums[0] < nums[1]:
            return nums[0]
        
        # invariance: single digits location is < hi and > lo, so hi - lo >= 2
        
        while True:
            mid = lo + int((hi- lo)/2)
            if nums[mid] < nums[mid + 1] and nums[mid] > nums[mid - 1]:
                return nums[mid]
            
            if nums[mid] == nums[mid + 1]:
                if hi > mid + 1 and (hi - mid -1) % 2 ==1:
                    lo = mid + 1
                else:
                    hi = mid
                
            else:
                # in this case, nums[mid] == nums[mid - 1]
                if hi > mid  and (hi - mid) % 2 ==1:
                    lo = mid
                else:
                    hi = mid
                
    def singleNonDuplicate(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """
        n = len(nums)
        if n == 1:
            return nums[0]
        lo = 0
        hi = n - 1
        while True:
            mid = (lo + hi) // 2
            has_double = False
            if mid != 0:
                has_double |= nums[mid - 1] == nums[mid]
            if mid != n - 1:
                has_double |= nums[mid + 1] == nums[mid]
            if not has_double:
                return nums[mid]
            odd_or_even = (mid - lo) % 2
            behind_matches = nums[mid] == nums[mid - 1]
            if odd_or_even:
                if behind_matches:
                    lo = mid + 1
                else:
                    hi = mid - 1
            else:
                if behind_matches:
                    hi = mid
                else:
                    lo = mid
        
            
```
