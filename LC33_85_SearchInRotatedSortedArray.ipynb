```python
class Solution:
    def binarySearch(self, lo, hi, nums, target):
        while lo<= hi:
            mid = lo + int((hi-lo)/2)
            if nums[mid] == target:
                return mid
            if nums[mid] > target:
                hi = mid - 1
            else:
                lo = mid + 1                        
        return -1
    
    def findMinIndex(self, nums):
        """
        :type nums: List[int]
        :rtype: int
        """

        # corner cases
        if len(nums) == 1 or nums[0] < nums[-1]:
            return 0

        if nums[-1] < nums[-2]:
            return len(nums) - 1

        # binary search idea
        # If we are here, array is at least length 3
        
        # Invariance we are maintaining: minimal number always at position > lo and < hi 

        lo = 0
        hi = len(nums) - 1
        while True:
            mid = lo + int((hi - lo)/2)
            if nums[mid] < nums[mid+1] and nums[mid] < nums[mid-1]:
                return mid
            if nums[mid] > nums[lo]:
                lo = mid
            else:
                hi = mid
    
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        # corner case
        if len(nums)==0:
            return -1
        
        # we first use log(n) time to find the minimal index
        # this requires nums to have length >= 1        
        minIdx = self.findMinIndex(nums)
        
        # And then we can do two binary search to find the item in the two sorted subarray 
        if minIdx == 0:            
            return self.binarySearch(0, len(nums)-1, nums, target)
        
#         if minIdx == len(nums) - 1:
#             if target == nums[-1]:
#                 return len(nums) - 1
#             return self.binarySearch(nums[:-1], target)
        
#         # if we are here, the min index is in the middle of the array, we need two binary searches
#         if minIdx == len(nums)
        
        rhs_search = self.binarySearch(minIdx, len(nums)-1, nums, target)
        if rhs_search != -1:
            return rhs_search 
        
        return self.binarySearch(0, minIdx - 1, nums, target)
        
```

The following is a simpler solution using binary search 

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        lo = 0
        hi = len(nums) - 1
        while lo <= hi:
            mid = lo + int((hi-lo)/2)
            if nums[mid] == target:
                return mid

            # if we are here, one of the right or left halves must be sorted (they can both be sorted)
            if nums[lo] <= nums[mid]: # since there are no duplicate, <= is the same as < here, 
                #however, for the corner case of two element array where lo == mid, we use <= so that we will gete to lo = mid +1 case when we have arr = [a1, a2] and a1 != target
                
                # sorted array must be the left half!
                if nums[lo] <= target < nums[mid]:
                    hi = mid - 1
                else:
                    lo = mid + 1
            else:
                # right half must be sorted! 
                if nums[mid] < target <= nums[hi]:
                    lo = mid + 1
                else:
                    hi = mid - 1
        return -1
```

Below is a solution modified from the binary search solution for search in rotated sorted array I

```python
class Solution(object):
    def search(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: int
        """
        lo = 0
        hi = len(nums) - 1
        while lo <= hi:
            mid = lo + int((hi-lo)/2)
            if nums[mid] == target:
                return True
            
            
            # the only difference with search in rotetad sorted array 1:
            # if nums[lo] <= nums[mid], we cannot conclude that left array is sorted and just go for the right half if target is not in the range of left array
            if nums[lo] == nums[mid] == nums[hi]:
                lo += 1
                hi -= 1
                continue
            
            # if we are here, one of the right or left halves must be sorted (they can both be sorted)
            if nums[lo] <= nums[mid]: # since there are no duplicate, <= is the same as < here, 
                #however, for the corner case of two element array where lo == mid, we use <= so that we will gete to lo = mid +1 case when we have arr = [a1, a2] and a1 != target
                
                # sorted array must be the left half!
                if nums[lo] <= target < nums[mid]:
                    hi = mid - 1
                else:
                    lo = mid + 1
            else:
                # right half must be sorted! 
                if nums[mid] < target <= nums[hi]:
                    lo = mid + 1
                else:
                    hi = mid - 1
        return False
```
